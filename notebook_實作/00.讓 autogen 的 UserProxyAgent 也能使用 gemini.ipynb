```python
# https://medium.com/@zelarsoft/using-google-gemini-pro-with-autogen-agents-08b55bba69db
# 看看是否讓 autogen 的 UserProxyAgent 可以使用 gemini-pro，請先安裝
# !pip install git+https://github.com/microsoft/autogen.git@gemini
# 上面是指定要安装的软件包的来源。它告诉pip从GitHub上的指定存储库克隆（clone）代码。git+告诉pip使用git协议，
# 后面的https://github.com/microsoft/autogen.git是GitHub存储库的网址，@gemini指定了git分支（branch），这表示你要安装的是gemini分支上的版本。
# 但是安裝gemini 分支，autogen 必須降版本到 v0.2.10，才能使用 gemini，如果使用最新版本 v0.2.15 autogenstudio 會報錯。
# autogenstudio 測試也可以叫得到了~~~~ 但 workfolw 設為 gemini 之後，一執行就馬上中斷，沒有對話能力????
# gemini 真的還不行，狀況比 ollama 還多
# 但是，雖然 autogenstudio 可以使用 ollama、LM Studio(gemini 當作還是不支援)，可是 autogen 又沒看到調用 ollama、LM Studio 本地大模型的案例
import autogen
from autogen import AssistantAgent, Agent, UserProxyAgent, ConversableAgent
from autogen.code_utils import DEFAULT_MODEL, UNKNOWN, content_str, execute_code, extract_code, infer_lang
```

```python
config_list = autogen.config_list_from_json(
    "OAI_CONFIG_LIST",
    filter_dict={
        # "model": ["gemini-pro"],
        "model": ["gpt-3.5-turbo-1106"],
    },
)
assistant = AssistantAgent(
    "assistant", 
    llm_config={"config_list": config_list_gemini, "seed": 42}, 
    max_consecutive_auto_reply=10
)

# 不可以設定 llm_config，不然會報錯，但沒有設好像也有可能彙報錯
user_proxy = UserProxyAgent(
    "user_proxy",
    code_execution_config={"work_dir": "coding", "use_docker": False},
    human_input_mode="NEVER",
    is_termination_msg=lambda x: content_str(x.get("content")).find("TERMINATE") >= 0,
)
```

```python
# 開始對話
# Plot a chart of NVDA and TESLA stock price change YTD
# 搜尋arXiv上有關醫學領域蛋白質研究最前緣的技術及連結(以markdown輸出，要注意避免給出重複的連結)，並評估哪一方面是最值得研究的領域
# Search arXiv for the most cutting-edge technologies and links related to protein research in the medical field (output in markdown, be careful to avoid giving duplicate links), and evaluate which aspect is the most worthy of research.
user_proxy.initiate_chat(assistant, message="""
    Search arXiv for the most cutting-edge technologies and links related to protein research in the medical field (output in markdown, be careful to avoid giving duplicate links), and evaluate which aspect is the most worthy of research.
""")
```
